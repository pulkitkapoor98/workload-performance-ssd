```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df1 = pd.read_csv('./results/RequestScale/ncq_depth_against.txt', sep=";")
df2 = pd.read_csv('./results/RequestScale/ncq_depth_favour.txt', sep=";")

dfs = pd.concat([df1['pre_depth'], df2['pre_depth']],axis=1, keys=['Against', 'Favour'])
ax = dfs.boxplot(grid=False, showfliers=False)
ax.set_xlabel("Workload")
ax.set_ylabel("NCQ Depth")
plt.title("Request Scale")
plt.show()
```

```python
import os
import json
from pprint import pprint
path = "./results/Locality/linux-dd-locality/"
cache_coverage = []
miss_ratio = []
for root, dirs, files in os.walk(path):
    for name in dirs:
        data = json.load(open(path+name+'/config.json'))
        cache = float(data["exp_parameters"]["cache_mapped_data_bytes"])/data["exp_parameters"]["lbabytes"]
        cache_coverage.append(round(cache, 2))
        data = json.load(open(path+name+'/recorder.json'))
        if('miss' in data["general_accumulator"]["Mapping_Cache"]):
            missRatio = float(data["general_accumulator"]["Mapping_Cache"]["miss"])/(int(data["general_accumulator"]["Mapping_Cache"]["miss"]) + int(data["general_accumulator"]["Mapping_Cache"]["hit"]))
        else:
            missRatio = 0
        miss_ratio.append(missRatio)

miss_ratio = [x for _,x in sorted(zip(cache_coverage,miss_ratio))]
cache_coverage.sort()
print cache_coverage,miss_ratio
fig, ax = plt.subplots()
ax.plot(cache_coverage, miss_ratio,'-')

ax.set(xlabel='Cache Coverage', ylabel='Miss Ratio',
       title='Locality')

plt.show()
```
